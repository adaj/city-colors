# city-colors

This projects aims to detect how much green area has in a GPS photo of the city and may helps to featurize city configuration with color clusters. The code implements the following steps:

* image ingestion and preprocessing 
* image color clustering using K-Means (K search optimized with silhouette coefficient)
* colors' percentage calculation
* semi-supervised color classification (to detect green)
* green area calculation and demonstration

### 0. Importing dependencies

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 

import os
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import cv2

from sklearn.linear_model import LogisticRegression
from sklearn.grid_search import GridSearchCV
import joblib
```

### 1. Image ingestion and preprocessing

```python
def load_image(path, show=False):
    # load the image and convert it from BGR to RGB so that
    # we can dispaly it with matplotlib
    image = cv2.imread(path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    if show:
        plt.figure()
        plt.imshow(image)
        plt.show();

    # reshape the image to be a list of pixels
    image = image.reshape((image.shape[0] * image.shape[1], 3))
    return image

image = load_image('photos/b600_zoom17.png',show=True)
```

### 2. Image color clustering using K-Means

```python
def image_kcolors(image, k=10, verbose=False):
    if k=='auto':
        # defining range_k as the parameter space to train K-Means
        range_k = range(4,9,2)
        s = []
        for k in range_k:
            km = KMeans(n_clusters = k, random_state=0)
            labels = km.fit_predict(image)

            # silhouette_scores perform K optimization
            silhouette_avg = silhouette_score(image, labels,sample_size=10000)
            s.append(silhouette_avg)
            if verbose:
                print('k={}  silhouette_score={}'.format(k,silhouette_avg))

        # picking k with the highest silhouette score
        k = range_k[s.index(max(s))]
    km = KMeans(n_clusters = k, random_state=0)
    km.fit(image)
    centroids = km.cluster_centers_
    return km, centroids

# km, centroids = image_kcolors(image, k='auto', verbose=True)
km, centroids = image_kcolors(image, k=10, verbose=True)
print('kcolors = \n',centroids)
```

### 3. Colors' percentage calculation

```python
def kcolors_pct(km, centroids, show=False):
    # grab the number of different clusters and create a histogram
    # based on the number of pixels assigned to each cluster
    numLabels = np.arange(0, len(np.unique(km.labels_)) + 1)
    (hist, _) = np.histogram(km.labels_, bins = numLabels)
    hist = hist.astype("float")
    hist /= hist.sum()

    if show:
        # initialize the bar chart representing the relative frequency
        # of each of the colors
        bar = np.zeros((50, 300, 3), dtype = "uint8")
        startX = 0
        
        # loop over the percentage of each cluster and the color of
        # each cluster
        for (percent, color) in zip(hist, centroids):
            # plot the relative percentage of each cluster
            endX = startX + (percent * 300)
            cv2.rectangle(bar, (int(startX), 0), (int(endX), 50),
                color.astype("uint8").tolist(), -1)
            startX = endX
            
        plt.figure()
        plt.axis("off")
        plt.imshow(bar)
        plt.show()
    return hist

kcolors_pct(km, centroids)
```

### 4. Semi-supervised color classification

This is where we have already clustered the colors, but the algorithm doesn't know what is considered 'green' for us. Then, it's adopted a strategy to teach it, labeling the colors from a training set of images to future use in a classifier. This process of labeling a small set of data train a classifier is called semi-supervised learning and it's a very used technique to learn in unlabeled data.

```python
import time
t0 = time.time()

path = 'photos/'

for i, photofile in enumerate(os.listdir(path)):
    if photofile.endswith('.png'):
        print('Processing {}'.format(path+photofile))
        kmeans, centroids = image_kcolors(load_image(path+photofile))
        if i==0:
            colors=np.concatenate((centroids,np.full((centroids.shape[0],1),photofile[0])),axis=1)
        else:
            colors=np.concatenate((colors,np.concatenate((centroids,np.full((centroids.shape[0],1),photofile[0])),axis=1)),axis=0)
print('Training data generated in {} seconds'.format(time.time()-t0))
            
dfcolors = pd.DataFrame(colors)
dfcolors['label'] = np.zeros(len(dfcolors))
dfcolors.head()
```

After we created the unlabeled dataset with clustered colors, let's run on each row of it and change 'label' column to 1 where we identify a green one. To do so, change **i** variable until the end of the dataframe and uncomment the last line if you see a green circle.

```python
colors = dfcolors[dfcolors.columns[0:3]].apply(lambda x: pd.to_numeric(x)).values

#iterate i manually to label the color as 1='green' and 0='not green'
i=0

img = np.zeros((512,512,3), np.uint8)
cv2.circle(img,(250,250), 200, colors[i], -1)
print(i,colors[i])
plt.imshow(img)
plt.show()

# dfcolors.loc[i,'label'] = 1     #### uncomment this line if colors[i] if green or grey-green
```

Ok, now let's train our classifier!

```python
def train_classifier(dfcolors):
    X, y = dfcolors[dfcolors.columns[0:3]], dfcolors['label']

    grid = {
            'C': np.power(10.0, np.arange(-10, 10)),
            'solver': ['newton-cg','lbfgs','sag']
        }
    clf = LogisticRegression(random_state=0, max_iter=10000, tol=5)
    classifier = GridSearchCV(clf, grid, scoring='accuracy', cv=5)
    classifier.fit(X, y)
    print ('gs.best_score_:', classifier.best_score_)

    joblib.dump(classifier, 'clf.pkl')
    return classifier

def predict_green(path, classifier):
    image = load_image(path, show=True)
    km, centroids = image_kcolors(image, k=10)
    pcts = kcolors_pct(km, centroids, show=True)
    ix = np.where(classifier.predict(centroids)==1)
    return np.sum([pcts[i] for i in ix])

classifier = train_classifier(dfcolors)
```
